```python
def run(cmds, num, cache, debug=False):
    _in = 0 
    i=0
    while num[i] == cache["last"][i]:
        i+=1
    #print(cache)
    cache["last"] = num
    mem = cache[i]
    for j,cmd in enumerate(cmds[cache[str(i)]:]):
        c = cmd[0]
        if c == "inp":
            if i == 1:
                return mem["z"]
            mem[cmd[1]] = int(num[i])
            cache[i] = {k: v for k,v in mem.items()}
            i += 1
            if debug:
                print(f"{cmd[1]} = {int(num[_in])}")
            continue
        else:
            if cmd[2] in "wxyz":
                v = mem[cmd[2]]
            else:
                v = int(cmd[2])
        if c == "add":
            mem[cmd[1]] = mem[cmd[1]]+v
            if debug:
                print(f"{cmd[1]} = {cmd[1]} + {cmd[2]} ({v})  ({mem[cmd[1]]})")
        elif c == "mul":
            mem[cmd[1]] = mem[cmd[1]]*v
            if debug:
                print(f"{cmd[1]} = {cmd[1]} * {cmd[2]} ({v})  ({mem[cmd[1]]})")
        elif c == "div":
            mem[cmd[1]] = mem[cmd[1]]//v
            if debug:
                print(f"{cmd[1]} = {cmd[1]} / {cmd[2]} ({v})  ({mem[cmd[1]]})")
        elif c == "mod":
            if mem[cmd[1]]<0 or v<=0:
                continue
            mem[cmd[1]] = mem[cmd[1]]%v
            if debug:
                print(f"{cmd[1]} = {cmd[1]} % {cmd[2]} ({v})  ({mem[cmd[1]]})")
        elif c == "eql":
            mem[cmd[1]] = 1 if mem[cmd[1]]==v else 0
            if debug:
                print(f"{cmd[1]} = {cmd[1]} == {cmd[2]} ({v})  ({mem[cmd[1]]})")
        else:
            print("ERROR")
    return mem["z"]        
    
def solution_a(data):
    cache = {"last":"0000000000000"}
    cache[0] = {"w":0,"x":0,"y":0,"z":0}
    inp = 0
    for i, cmd in enumerate(data):
        if cmd[0][:3] == "inp":
            cache[str(inp)] = i
            inp += 1
    x="99999999999999"
    m=9999999999999
    for h in range(99999999999999,11111111111111,-1):
        s = str(h)
        if s.find("0") > -1:
            continue
        r = run(data,s, cache)
        if r < m:
            m = r
            print(f"{s} -> {m}")
        
    return "0"
```
